```python
import pandas as pp
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

data=pp.read_csv("study.csv")
data

```

```python

X= data.drop(columns=['exam_score'])
y=data['exam_score']

model=DecisionTreeClassifier()
model.fit(X,y)
```

```python
tree.export_graphviz(model,out_file='exam.dot',feature_names=['study_hours','sleep_hours','practice_hours'],
                     label='all',
                     rounded=True,
                     filled=True)
```
